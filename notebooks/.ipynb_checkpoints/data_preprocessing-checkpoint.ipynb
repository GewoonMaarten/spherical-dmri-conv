```python
from nilearn.masking import apply_mask
from nilearn import image, plotting
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_file = 'MB_Re_t_moco_registered_applytopup.nii.gz'
msk_file = 'brain_mask.nii.gz'
```

```python
direc11 = './data/cdmri0011/'
selected_imgs = image.index_img(os.path.join(direc11, img_file), np.array([1, 10, 100, 1000]))
for img in image.iter_img(selected_imgs):
    # img is now an in-memory 3D img
    plotting.plot_anat(img)
```

```python
plotting.plot_anat(os.path.join(direc11, msk_file))
```

```python
direc11 = './data/cdmri0011/'
masked_data11 = np.transpose(apply_mask(imgs=os.path.join(direc11, img_file),
                                        mask_img=os.path.join(direc11, msk_file)))
print(masked_data11.shape)
```

```python
direc12 = './data/cdmri0012/'
masked_data12 = np.transpose(apply_mask(imgs=os.path.join(direc12, img_file),
                                        mask_img=os.path.join(direc12, msk_file)))
```

```python
direc13 = './data/cdmri0013/'
masked_data13 = np.transpose(apply_mask(imgs=os.path.join(direc13, img_file),
                                        mask_img=os.path.join(direc13, msk_file)))
```

```python
direc14 = './data/cdmri0014/'
masked_data14 = np.transpose(apply_mask(imgs=os.path.join(direc14, img_file),
                                        mask_img=os.path.join(direc14, msk_file)))
```

```python
direc15 = './data/cdmri0015/'
masked_data15 = np.transpose(apply_mask(imgs=os.path.join(direc15, img_file),
                                        mask_img=os.path.join(direc15, msk_file)))
```

```python
max_data = np.percentile(masked_data11.flatten(), 95)
print(max_data)
```

```python
fig, axes = plt.subplots(1, 5, sharey=True, figsize=(20,10))
axes[0].hist(masked_data11.flatten(), range=[0, max_data])
axes[1].hist(masked_data12.flatten(), range=[0, max_data])
axes[2].hist(masked_data13.flatten(), range=[0, max_data])
axes[3].hist(masked_data14.flatten(), range=[0, max_data])
axes[4].hist(masked_data15.flatten(), range=[0, max_data])
fig.show()
```

# Median across all images

```python
med11 = np.median(masked_data11, axis=0)
med12 = np.median(masked_data12, axis=0)
med13 = np.median(masked_data13, axis=0)
med14 = np.median(masked_data14, axis=0)
med15 = np.median(masked_data15, axis=0)
```

```python
a12, _, _, _ = np.linalg.lstsq(med12[:,np.newaxis],med11)
a13, _, _, _ = np.linalg.lstsq(med13[:,np.newaxis],med11)
a14, _, _, _ = np.linalg.lstsq(med14[:,np.newaxis],med11)
a15, _, _, _ = np.linalg.lstsq(med15[:,np.newaxis],med11)
print(a12,a13,a14,a15)
```

```python
fig, axes = plt.subplots(1, 4, sharey=True, figsize=(20,5))
axes[0].plot(med12,med11, 'yo', med12, med12*a12, ':k', med12, med12, '-k')
axes[1].plot(med13,med11, 'bo', med13, med13*a13, ':k', med13, med13, '-k')
axes[2].plot(med14,med11, 'ro', med14, med14*a14, ':k', med14, med14, '-k')
axes[3].plot(med15,med11, 'go', med15, med15*a15, ':k', med15, med15, '-k')
for ax in axes:
    ax.set(aspect='equal')
fig.show()
```
